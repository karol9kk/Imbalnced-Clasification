```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generowanie przykadowych danych

x = np.linspace(-75, 100, 20).reshape(-1, 1)
y = 5*x**5 - 8*x**4 + 10*x**3 - 12*x**2 + 15*x - 20 + np.random.randint(7) # Poprawka

# Tworzenie rnych stopni wielomianw
degrees = [1, 2, 3, 4, 5]
plt.figure(figsize=(10, 6))

# Wywietlenie punktw
plt.scatter(x, y, color='black')

for degree in degrees:
    # Przygotowanie danych wielomianowych
    polynomial_features = PolynomialFeatures(degree=degree)
    x_poly = polynomial_features.fit_transform(x)
    
    # Regresja liniowa
    model = LinearRegression()
    model.fit(x_poly, y)
    
    # Przewidywanie wartoci dla nowych danych
    x_new = np.linspace(-75, 100, 20).reshape(-1, 1)
    x_new_poly = polynomial_features.transform(x_new)
    y_new_poly_pred = model.predict(x_new_poly)
    
    # Wywietlenie krzywej regresji
    plt.plot(x_new, y_new_poly_pred, label='Stopie {}'.format(degree))

plt.title('Regresja wielomianowa dla rnych stopni wielomianw')
plt.xlabel('X')
plt.ylabel('Y')
plt.xscale('linear')  
plt.yscale('linear')  
plt.legend()
plt.grid(True)
plt.show()
```
