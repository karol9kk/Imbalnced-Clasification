```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import KMeansSMOTE
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import EditedNearestNeighbours
from imblearn.under_sampling import RandomUnderSampler


sampling=RandomOverSampler(random_state=42,sampling_strategy=1)
#under=EditedNearestNeighbours(n_neighbors=2,kind_sel='mode')
#under=RandomUnderSampler(sampling_strategy=1)
sampling=KMeansSMOTE(sampling_strategy=0.9,random_state=42,n_jobs=-1,k_neighbors=4,cluster_balance_threshold=0.001)
#sampling=SMOTE(sampling_strategy=0.5,random_state=42)

train_features,train_labels=sampling.fit_resample(train_features,train_labels)
```

```python
import tensorflow as tf
from tensorboard.plugins.hparams import api as hp
from tensorflow import keras

# Define your hyperparameters
HP_NUM_UNITS_1 = hp.HParam('num_units_1', hp.Discrete([32,64, 128]))
HP_NUM_UNITS_2 = hp.HParam('num_units_2', hp.Discrete([32,64, 128]))
HP_NUM_UNITS_3 = hp.HParam('num_units_3', hp.Discrete([32, 64]))
HP_DROPOUT = hp.HParam('dropout', hp.RealInterval(0.1, 0.2))

# Define the model
def make_model(hparams):
    model = keras.Sequential([
        keras.layers.Dense(hparams[HP_NUM_UNITS_1], activation='relu',kernel_regularizer='l2',
                           input_shape=(train_features.shape[-1],)),
        keras.layers.Dropout(hparams[HP_DROPOUT]),
        keras.layers.Dense(hparams[HP_NUM_UNITS_2], activation='relu',kernel_regularizer='l2'),
        keras.layers.Dropout(hparams[HP_DROPOUT]),
        keras.layers.Dense(hparams[HP_NUM_UNITS_3], activation='relu',kernel_regularizer='l2'),
        keras.layers.Dropout(hparams[HP_DROPOUT]),
        keras.layers.Dense(1, activation='sigmoid'),
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.01),
        loss=keras.losses.BinaryCrossentropy(),
        metrics=['accuracy']
    )
    
    return model

# Define the function to run a single training session
def run(run_dir, hparams):
    with tf.summary.create_file_writer(run_dir).as_default():
        hp.hparams(hparams) 
        model = make_model(hparams)
        model.fit(train_features, train_labels, epochs=EPOCHS, batch_size=BATCH_SIZE, callbacks=[early_stopping])
        loss = model.evaluate(test_features, test_labels)[0]
        tf.summary.scalar('loss', loss, step=1)

# Set your epochs and batch size
EPOCHS = 15
BATCH_SIZE = 256

# Set up early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='loss', 
    verbose=1,
    patience=25,
    restore_best_weights=True)

# Set up hyperparameter search
session_num = 0

for num_units_1 in HP_NUM_UNITS_1.domain.values:
    for num_units_2 in HP_NUM_UNITS_2.domain.values:
        for num_units_3 in HP_NUM_UNITS_3.domain.values:
            for dropout_rate in (HP_DROPOUT.domain.min_value, HP_DROPOUT.domain.max_value):
                hparams = {
                    HP_NUM_UNITS_1: num_units_1,
                    HP_NUM_UNITS_2: num_units_2,
                    HP_NUM_UNITS_3: num_units_3,
                    HP_DROPOUT: dropout_rate,
                }
                run_name = "run-%d" % session_num
                print('--- Starting trial: %s' % run_name)
                print({h.name: hparams[h] for h in hparams})
                run('logs/hparam_tuning/' + run_name, hparams)
                session_num += 1
```
